```python
def Ustat(K,y,Samp1,Samp2):
    U = 0.
    Usum = 0.
    for i in Samp1:
        for j in Samp2:
            if K[i,j] > 0:
                Usum += 1
                if y[i] > y[j]:
                    U += 1
    return U, Usum
```

```python
def Ustat(K,y,Samp1,Samp2):
    n = K.shape[0]
    Y = y.reshape((n,1))
    KY = K * (Y - Y.T > 0)
    UMat = KY[Samp1,:][:,Samp2]
    SMat = K[Samp1,:][:,Samp2]
    return UMat.sum(), SMat.sum()
```

```python
U, Usum = Ustat(K,TMP,Premier,NPremier)
```

```python
def permStat():
    Samp1, Samp2 = randSamp(n,s)
    Umc, Umcsum = Ustat(K,TMP,Samp1,Samp2)
    return Umc - Umcsum/2
```
